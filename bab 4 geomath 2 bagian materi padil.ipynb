## 4.1.1 Transformasi Laplace dari Fungsi Tangga Satuan

Fungsi tangga satuan atau fungsi langkah Heaviside didefinisikan sebagai
$$
u_s(t) =
\begin{cases}
1 & \text{for } t \ge 0 \\
0 & \text{for } t < 0
\end{cases}
\quad (4.1.2)
$$

Transformasi Laplace dari fungsi ini $(\mathcal{L}\{u_s(t)\})(s)$
dihitung menggunakan definisi integral transformasi Laplace:
$$
\mathcal{L}\{u_s(t)\} = \int_0^\infty u_s(t) e^{-st} dt
$$
Karena $u_s(t) = 1$ untuk $t \ge 0$, integral menjadi:

$$
\mathcal{L}\{u_s(t)\} = \int_0^\infty 1 \cdot e^{-st} dt
$$

Menyelesaikan integral:

$$
\mathcal{L}\{u_s(t)\} = \left[ \frac{e^{-st}}{-s} \right]_{0}^{\infty}
$$

Mensubstitusikan batas-batas:

$$
\mathcal{L}\{u_s(t)\} = \lim_{t \to \infty} \left( \frac{e^{-st}}{-s} \right) - \left( \frac{e^{-s(0)}}{-s} \right)
$$

Asumsikan $s > 0$ agar batas tak hingga menjadi nol:

$$
\mathcal{L}\{u_s(t)\} = (0) - \left(\frac{1}{-s}\right) = \frac{1}{s}
$$

**Pasangan Transformasi Laplace** ini adalah:

$$
u_s(t) \;\;\longleftrightarrow\;\; \frac{1}{s}
$$


## 4.1.2 Transformasi Laplace dari Fungsi Impuls Satuan

*Impuls satuan* atau *fungsi delta Dirac* dapat didefinisikan sebagai limit dari pulsa persegi dengan luas satu (unit-area rectangular pulse function) 
$\frac{r_D(t)}{D}$ ($r_D(t)$: ditunjukkan pada Gambar 4.1(b)) dengan lebar pulsa $D \to 0$ (mengecil hingga nol) atau secara sederhana, sebagai turunan terhadap waktu dari fungsi langkah satuan (unit step function):

$$
\delta(t) = \lim_{D \to 0} \frac{1}{D} r_D(t) 
= \lim_{D \to 0} \frac{1}{D} \Big(u_s(t + \tfrac{D}{2}) - u_s(t - \tfrac{D}{2})\Big) 
= \frac{d}{dt} u_s(t) \tag{4.1.4}
$$

Dengan menggunakan Pers. (4.1.1), kita dapatkan transformasi Laplace dari fungsi impuls satuan sebagai berikut:

$$
\mathcal{L}\{\delta(t)\} = \int_{0^-}^{\infty} \delta(t) e^{-st} dt 
= e^{-s \cdot 0} \Big|_{t=0} \int_{0^-}^{\infty} \delta(t) dt^{\text{ luas = 1}} = 1
\tag{4.1.5}
$$

Pasangan transformasi Laplace ini dapat dituliskan:

$$
\delta(t) \;\;\xleftrightarrow{\;\;\mathcal{L}\;\;} 1
$$

Persamaan (4.1.4) menyatakan bahwa integral dari fungsi impuls satuan adalah fungsi langkah satuan:

$$
\int_{-\infty}^{t} \delta(\tau) \, d\tau = u_s(t) \tag{4.1.6}
$$


### Catatan 4.1
**Sifat-sifat Fungsi Impuls Satuan** $\delta(t)$  

1. Fungsi impuls satuan $\delta(t)$ memiliki luas sama dengan satu di sekitar $t = 0$, yang berarti:

$$
\int_{-\infty}^{+\infty} \delta(\tau)\, d\tau = \int_{0^-}^{0^+} \delta(\tau)\, d\tau = 1 \tag{4.1.7}
$$

Atau secara ekuivalen:

$$
\left\{ \; \int_{-\infty}^{0^+} \delta(\tau)\, d\tau - \int_{-\infty}^{0^-} \delta(\tau)\, d\tau 
= u_s(0^+) - u_s(0^-) = 1 - 0 = 1 \;\right\}
$$


(2) Fungsi impuls satuan $\delta(t)$ bersifat simetris terhadap $t=0$:

$$
\delta(t) = \delta(-t) \tag{4.1.8}
$$



(3) Konvolusi dari suatu fungsi $x(t)$ dengan fungsi impuls satuan $\delta(t)$ menghasilkan fungsi itu sendiri:

$$
x(t) * \delta(t) 
\;\;\overset{\text{definisi integral konvolusi}}{=}\;\;
\int_{-\infty}^{+\infty} x(\tau)\delta(t-\tau)\, d\tau = x(t) \tag{4.1.9}
$$



Bagaimana dengan konvolusi fungsi waktu $x(t)$ dengan impuls satuan yang tertunda $\delta(t-t_1)$?  
Hasilnya adalah fungsi yang tertunda atau bergeser $x(t-t_1)$:

$$
x(t) * \delta(t-t_1) 
= \int_{-\infty}^{+\infty} x(\tau)\delta(t-\tau-t_1)\, d\tau 
= x(t-t_1) \tag{4.1.10}
$$



Bagaimana dengan konvolusi fungsi waktu yang sudah tertunda $x(t-t_2)$ dengan impuls satuan tertunda $\delta(t-t_1)$?  
Hasilnya adalah fungsi yang tertunda lebih jauh $x(t-(t_1+t_2))$:

$$
x(t-t_2) * \delta(t-t_1) 
= \int_{-\infty}^{+\infty} x(\tau - t_2)\delta(t-\tau-t_1)\, d\tau 
= x(t-(t_1+t_2)) \tag{4.1.11}
$$



Jika $x(t) * y(t) = z(t)$, maka berlaku

$$
x(t-t_1) * y(t-t_2) = z(t-(t_1+t_2)) \tag{4.1.12}
$$

Namun, jika $t$ diganti dengan $t-t_1$ pada kedua sisi, hasilnya tidak berlaku, yaitu:

$$
x(t-t_1) * y(t-t_1) \neq z(t-t_1), \quad 
x(t-t_1) * y(t-t_1) = z(t-2t_1)
$$



(4) Fungsi impuls satuan $\delta(t)$ memiliki **sifat sampling atau penyaringan (sifting property)**:

$$
\int_{-\infty}^{+\infty} x(t)\delta(t-t_1)\, dt 
= x(t_1) \tag{4.1.13}
$$

Sifat ini memungkinkan kita untuk “mengambil sampel” nilai dari sinyal kontinu $x(t)$ pada saat tertentu $t=t_1$.  
Sifat ini juga bisa digunakan untuk memodelkan sinyal waktu-diskrit yang diperoleh dari proses sampling sinyal waktu-kontinu.


## 4.2.1 Linearitas

Kelinearan adalah sifat mendasar yang menyatakan bahwa Transformasi Laplace dari kombinasi linear dua fungsi waktu $(αx(t)+βy(t))$ sama dengan kombinasi linear dari Transformasi Laplace masing-masing fungsi tersebut $(αX(s)+βY(s))$.
$$
\alpha x(t) + \beta y(t) 
\;\xleftrightarrow{\mathcal{L}}\;
\alpha X(s) + \beta Y(s)
\tag{4.2.1}
$$
Yang mana,
$$
\begin{aligned}
& \alpha \text{ dan } \beta = \text{konstanta (bisa bilangan real atau kompleks)} \\
& X(s) = \mathcal{L}\{x(t)\} \\
& Y(s) = \mathcal{L}\{y(t)\}
\end{aligned}
$$


Artinya, jika kita melakukan transformasi Laplace terhadap penjumlahan fungsi yang dikalikan konstanta, hasilnya akan sama dengan penjumlahan transformasi Laplace dari masing-masing fungsi yang juga dikalikan konstanta yang sama.




## 4.2.2 Diferensiasi Waktu (Time Differentiation)

Transformasi Laplace dari turunan$ $x(t)$ terhadap $(t)$ dapat ditulis sebagai
ransformasi Laplace memiliki sifat penting yang berkaitan dengan **turunan terhadap waktu**. Sifat ini sangat berguna karena mengubah operasi **turunan (yang sulit di domain waktu)** menjadi **perkalian sederhana dengan s** di domain Laplace.

$$
x'(t) \;\xleftrightarrow{\mathcal{L}}\; sX(s) - x(0)
\tag{4.2.2}
$$

**(Bukti)**

$$
\begin{aligned}
\mathcal{L}\{x'(t)\}
& \overset{(4.1.1)}{=}
\int_{0}^{\infty} \frac{dx}{dt} e^{-st} \, dt \\
& \overset{(C.38)}{=}
x(t)e^{-st} \Big|_{0}^{\infty}
- (-s)\int_{0}^{\infty} x(t)e^{-st} \, dt \\
& \overset{(4.1.1)}{=}
sX(s) - x(0)
\end{aligned}
$$



**282 Bab 4 Transformasi Laplace**

Penerapan berulang dari sifat diferensiasi waktu menghasilkan Transformasi Laplace dari turunan ke-$(n)$ dari $x(t)$ terhadap $(t)$ sebagai

$$
x^{(n)}(t)
\;\xleftrightarrow{\mathcal{L}}\;
s^n X(s)
- s^{n-1}x(0)
- s^{n-2}x'(0)
- \cdots -
x^{(n-1)}(0)
\tag{4.2.3}
$$


## 4.2.6 Konvolusi Riil
Konvolusi adalah operasi matematika yang menggambarkan bagaimana dua sinyal saling mempengaruhi dalam domain waktu.
Konvolusi (riil) dari dua fungsi (kausal) $g(t)$ dan $x(t)$ didefinisikan sebagai

$$
g(t) * x(t) = \int_{-\infty}^{\infty} g(\tau)x(t-\tau)\,d\tau
\quad
\underset{\text{jika } g(t)=0 \text{ dan } x(t)=0 \text{ untuk } t<0}{\xrightarrow{\text{kausalitas}}}
\quad
\int_0^t g(\tau)x(t-\tau)\,d\tau
\tag{4.2.8}
$$


Transformasi Laplace dari konvolusi \( y(t) = g(t) * x(t) \) ternyata merupakan hasil kali dari transformasi Laplace kedua fungsi tersebut, yaitu

$$
y(t) = g(t) * x(t)
\;\xleftrightarrow{\mathcal{L}}\;
Y(s) = G(s)X(s)
\tag{4.2.9}
$$



**(Pembuktian)**

$$
\begin{aligned}
\mathcal{L}\{g(t)*x(t)\}
& \overset{(4.1.1)}{=} 
\int_0^{\infty} [g(t)*x(t)] e^{-st}\,dt \\
& \overset{(4.2.8)}{=} 
\int_0^{\infty} \left[\int_0^{t} g(\tau)x(t-\tau)\,d\tau \right] e^{-st}\,dt \\
& \overset{\text{Gbr. 4.3}}{=} 
\int_0^{\infty} g(\tau)e^{-s\tau} \left[\int_0^{\infty} x(v)e^{-sv}\,dv \right] d\tau \\
& \overset{(4.1.1)}{=} 
\left[\int_0^{\infty} g(\tau)e^{-s\tau}\,d\tau \right]
\left[\int_0^{\infty} x(v)e^{-sv}\,dv \right] \\
& = G(s)X(s)
\end{aligned}
$$

Sifat ini digunakan untuk menggambarkan hubungan **masukan–keluaran** dari suatu **sistem waktu kontinu linear invarian (LTI)** yang memiliki masukan $x(t)$, **respon impuls** $g(t)$ (yakni keluaran akibat masukan impuls), dan keluaran $y(t)$, di mana transformasi Laplace dari respon impuls tersebut, yaitu $G(s) = \mathcal{L}\{g(t)\} $, disebut sebagai **fungsi sistem** atau **fungsi alih (transfer function)** dari sistem tersebut.
